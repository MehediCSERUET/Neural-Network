```python
#Code for Hopfield neural network
import numpy as np
import matplotlib.pyplot as plt
from numpy import binary_repr
import random
from sklearn.model_selection import train_test_split
import pandas as pd
np.random.seed(100)
```

```python
def multiply(x, i, j):
    ans = 0
    for k in range(len(x)):
        ans += x[k][i] * x[k][j]
    return ans
```

```python
def hard_limiting_threshold(x):
    #print(x)
    if x <= 0:
        return -1.0
    else:
        return 1.0
```

```python
def n_multiply(w, miu, nb_node):
    ans = []
    for i in range(nb_node):
        ans.append(hard_limiting_threshold(np.dot(np.array(w[i]).T, np.array(miu))))
    return ans
```

```python
def match(a, b):
    for i in range(len(a)):
        if round(a[i]) != round(b[i]):
            return False
    return True
```

```python
def hopfield(x):
    nb_node = len(x[0])
    w = np.zeros((nb_node, nb_node))
    for i in range(nb_node):
        for j in range(nb_node):
            if i is not j:
                w[i][j] = multiply(x, i, j)
    print(w)
    return w
```

```python
def test_hopfield(x, w):
    nb_node = len(x[0])
    for i in range(len(x)):
        print('\nFor input: ', x[i])
        miu = x[i]
        while True:
            tmp = miu
            miu = n_multiply(w, miu, nb_node)
            if match(miu , tmp):
                print('Predicted',miu)
                break
```

```python
if __name__=="__main__":
    train = pd.read_csv('Train-Data.csv').values
    test = pd.read_csv('Train-Data.csv').values
    
    #print('Train\n',train)
    #print('Test\n',test)
    w = hopfield(train)
    test_hopfield(test, w)
```
